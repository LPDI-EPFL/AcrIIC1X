```python
def plot_per_res(scorefile, cols,
                 zero_line=False,
                 left_label="Score",
                 right_label=None,
                 title="Per residue scores",
                 left_y_lim=None,
                 right_y_lim=None,
                 width=7.06,
                 height=1.4,
                 show_seq=False,
                 show_pos=True):
    
    
    df = pd.read_table(scorefile, delim_whitespace=True)
    if cols == ['rama_prepro']:
        df = df[df['rama_prepro'] != 0.]
    else:
        df = df[df['restype2'] == "onebody"]
    
    # from https://github.com/juzb/DeeProtein/blob/master/DeeProtein/scripts/plot_sensitivity.py
    seq = ''.join([three2single[aa] for aa in df['restype1']])
    lw=1
    
    start_index = 1
    end_index = len(seq)
   
    width_per_aa = 0.08
    base_width = 0.5
    
    if not width:
        width = base_width + width_per_aa*(end_index-start_index) 
    
    fig = plt.figure(figsize=(width, height), dpi=150)
    ax = fig.add_subplot(1, 1, 1)
        
    x_idx = np.arange(start_index, end_index + 1)

    for plot_idx, col in enumerate(cols):    
        if plot_idx == 1:
            ax.plot(x_idx, df[col], lw=lw, label=col) # put this on right y axis
        else:
            ax.plot(x_idx, df[col], lw=lw, label=col)

    
    # zero line
    if zero_line:
        ax.plot([min(x_idx), max(x_idx)], [0.0, 0.0], color="black", label="Baseline", lw=1)
    
    if left_y_lim:
        ax.set_ylim(left_y_lim)

    if show_seq:
        seq_letters = [l for l in seq]
    
        # TWO TICK LABELS LEVELS: (with sequence:)
        text_ticks = plt.FixedFormatter(seq_letters)
        ax.xaxis.set_minor_locator(plt.FixedLocator(x_idx))
        x_idx_major = [x for x in x_idx if (x-x_idx[0])%10 == 0]
        if show_pos:
            x_idx_major_labels = [x for x in x_idx_major]
            
        else:
            x_idx_major_labels = [None for x in x_idx_major]
        major_ticks = plt.FixedFormatter(x_idx_major_labels)
        ax.xaxis.set_minor_formatter(text_ticks)
        ax.xaxis.set_major_locator(plt.FixedLocator(x_idx_major))
        ax.xaxis.set_major_formatter(major_ticks)
    
    if show_pos:
    # ONLY POSITION:
        x_idx_minor = [1] + [x for x in x_idx if (x-(x_idx[0]-1))%20 == 0]
        x_idx_minor_labels = [x for x in x_idx_minor]
        ax.set_xticks(x_idx_minor)
        ax.set_xticklabels(x_idx_minor_labels )

    ax.set_ylabel(left_label)    
    ax.set_title(title)
    
    if show_seq:
        ax.tick_params(axis="x", which="major", length=0, pad=15)
    # resize
    fig.set_dpi(300)
    fig.set_size_inches(width, height, forward=True)  
    
    
    # legend figure:
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(bbox_to_anchor=(0.015, 1.01, 1., .101),
               columnspacing=1, 
               loc='upper left', 
               ncol=len(handles),
              )
    
    
    ax.set_xlim(1, end_index)


    return fig, df, seq
```
